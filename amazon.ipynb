# 📊 Amazon Product Data Analysis using Python

### 🔍 Analyzing product pricing, ratings, and trends using Pandas, Matplotlib, and Seaborn

## 📌 Introduction

This project focuses on analyzing Amazon product data to understand pricing, discounts, ratings, and customer behavior.  
The goal is to extract meaningful insights that can help in better decision-making.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("sales_dataset.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Clean price columns
df["discounted_price"] = df["discounted_price"].str.replace("₹", "").str.replace(",", "").astype(float)
df["actual_price"] = df["actual_price"].str.replace("₹", "").str.replace(",", "").astype(float)

# Clean discount percentage
df["discount_percentage"] = df["discount_percentage"].str.replace("%", "").astype(float)

df.head()
```

```python
# Total Revenue
total_revenue = df["discounted_price"].sum()
print("Total Revenue:", total_revenue)
```

### Explanation:
This calculates the total revenue by summing all discounted prices.

### Insight:
It shows the overall value of products after discount.

## 📊 Exploratory Data Analysis (EDA)

We analyze the dataset using various visualizations and calculations to identify patterns and trends.

### 📦 Category-wise Product Distribution

```python
df["category"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Categories by Product Count")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

### Explanation:
This graph shows the number of products in each category.

### Insight:
Categories with more products dominate the platform.

### 💰 Price Distribution Analysis

```python
sns.histplot(df["discounted_price"], bins=30)
plt.title("Price Distribution")
plt.show()
```

### Explanation:
This shows how product prices are distributed.

### Insight:
Most products fall within a common price range.

### ⭐ Rating Distribution

```python
sns.histplot(df["rating"], bins=20)
plt.title("Rating Distribution")
plt.show()
```

### Explanation:
This shows how customers rate products.

### Insight:
Most products have high ratings, indicating good satisfaction.

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
```

### 📉 Discount Percentage Analysis

```python
df.groupby("category")["rating"].mean().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Top Categories by Rating")
plt.show()
```

### Explanation:
This shows average ratings per category.

### Insight:
Some categories consistently receive better ratings.

### 🔥 Correlation Heatmap

```python
numeric_df = df[["discounted_price", "actual_price", "rating", "rating_count"]]

sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")
```

```python
df = df.dropna(subset=["discounted_price", "actual_price", "rating", "rating_count"])
```

### Explanation:
This shows relationships between numerical variables.

### Insight:
It helps identify how price and ratings are connected.

```python
sns.scatterplot(x="actual_price", y="discounted_price", data=df)
plt.title("Actual vs Discounted Price")
plt.show()
```

### Explanation:
This compares actual price and discounted price.

### Insight:
Higher priced products usually have higher discounts.
